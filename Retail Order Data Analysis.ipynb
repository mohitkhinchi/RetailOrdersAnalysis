```python
df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])

df['Ship Mode'].unique()
```

```python
# Correcting the column names: Make them lower case and replace with underscore

# df.rename(columns = {'Order Date': 'order_date', 'Ship Mode': 'ship_mode'})

df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')
```

```python
df.columns
```

```python
df.head()
```

```python
# Derive new columns "discount", "sale_price", "profit"

df['discount']= df['list_price']*df['discount_percent']*0.01

df['sale_price'] = df['list_price'] - df['discount']

df['profit'] = df.sale_price - df.cost_price

df.head()
```

```python
df.dtypes
```

```python
# Convert date coulmn to Object Datatype to DateTime

df['order_date']= pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

```python
df.dtypes
```

```python
# Drop non required columns

df.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace = True)

df.head()
```
